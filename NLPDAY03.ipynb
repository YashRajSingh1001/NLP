```python
#BoW implementation using NLTK
import pandas as pd
data = pd.read_csv("/content/SMSSpamCollection.txt", sep="\t",names=['label','message'])
data.head()
```

```python
#Data Cleanining and Preprocessing
import re
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
ps = PorterStemmer()
```

```python
corpus = []
for i in range(0,len(data)):
  review = re.sub('[^a-zA-Z]',' ',data['message'][i]) #removing special characters
  review = review.lower()
  review = review.split()
  review = [ps.stem(word) for word in review if not word in stopwords.words("english")]
  review = ' '.join(review)
  corpus.append(review)
```

```python
#Creating BoW Model
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=2500, binary = True)
X = cv.fit_transform(corpus).toarray()
X.shape
```

```python
cv.vocabulary_
```

```python
#N-grams - Eg: bigrams, trigrams

#S1 - the food is good ----> food, good (1,0,1)
#S2 - the food is not good ----> food, not, good (1,1,1)

#Not much of a difference, will try to sort this issue using ngrams

#If we use bigrams in this situation then it will make a pair of two words - (food, not, good, food good, not good, food not)
#(1,0,1,1,0,0)
#(1,1,1,0,1,1)

#Now, the model will be able to make a clear difference

#sklearn has this terminology -
#n-grams=(1,1) unigrams
#n-grams=(1,2) unigram, bigram
#n-grams=(1,3) bigram, trigram
```

```python
from nltk.stem import WordNetLemmatizer
nltk.download('wordnet')
lem = WordNetLemmatizer()
```

```python
corpus = []
for i in range(0,len(data)):
  review = re.sub('[^a-zA-Z]',' ',data['message'][i])
  review = review.lower()
  review = review.split()
  review = [lem.lemmatize(word) for word in review if not word in stopwords.words("english")]
  review = ' '.join(review)
  corpus.append(review)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
cv = TfidfVectorizer(max_features=100)
X = cv.fit_transform(corpus).toarray()
X.shape
```
